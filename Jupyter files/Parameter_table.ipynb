```python
import pandas as pd
import matplotlib.pyplot as plt

# Define parameters for PID control
parameters = [
    "ThrottlePID - Kp", "ThrottlePID - Ki", "ThrottlePID - Kd",
    "YawPID - Kp", "YawPID - Ki", "YawPID - Kd",
    "XposPID - Kp", "XposPID - Ki", "XposPID - Kd",
    "YposPID - Kp", "YposPID - Ki", "YposPID - Kd"
]

# Simulated values for Default, Tuned-Avg, and Global-Best
default_values = [4.0, 0.5, 0.3, 30.0, 10.0, 0.2, 0.2, 0.1, 0.1, 0.2, 0.1, 0.1]
tuned_avg_values = [2.5014500448835593, 14.272830803416374, 0.6717933057329903, 1.974266794376939, -2.1041968992493625, 4.683498378641889, 5.272423930276116, 1.0259514065410273, 3.4803685721900837, 2.4763544410198848, 1.4126215193096952, 4.732345420236985]
global_best_values = [2.3650709996437307, 15.783028433774358, 1.0303006395394594, 2.0411331287423815, -2.3840845374652018, 4.995190074976211, 5.161241146374655, 1.309654063028413, 3.6980714494167706, 2.142635042451028, 0.8522508279225341, 5.067396360736699]

# Create the table
data = {
    "Parameter": parameters,
    "Default": default_values,
    "Tuned-Avg": tuned_avg_values,
    "Global-Best": global_best_values
}

df = pd.DataFrame(data)

# Display the table
print("\nTABLE I: Final Particle Disparity for Applied PID Controls\n")
print(df)

# Plot performance metrics for altitude, yaw, X-position, and Y-position
performance_metrics = {
    "Altitude (Throttle)": [default_values[0], tuned_avg_values[0], global_best_values[0]],
    "Yaw": [default_values[3], tuned_avg_values[3], global_best_values[3]],
    "X-Position": [default_values[6], tuned_avg_values[6], global_best_values[6]],
    "Y-Position": [default_values[9], tuned_avg_values[9], global_best_values[9]],
}

# Create subplots for each metric
plt.figure(figsize=(14, 8))
for i, (metric, values) in enumerate(performance_metrics.items(), 1):
    plt.subplot(2, 2, i)
    plt.bar(["Default", "Tuned-Avg", "Global-Best"], values, alpha=0.7)
    plt.title(metric)
    plt.xlabel("PID Configurations")
    plt.ylabel("Value")
    plt.grid(axis="y")

plt.tight_layout()
plt.show()
```
